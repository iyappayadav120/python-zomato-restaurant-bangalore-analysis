```python
# Lets Drop the columns now 

df = df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1) 

# please note that i have dropped the columns as per my logic and my comfort which felt was not required

df.head()

# LEts check now the columns been dropped

# As you can see that all the above mentioned columns been dropped.
```

```python
# LEts check the unique values in rale column 

df['rate'].unique()

# As you can see in the output below there are 3 string values that are -NEW,'-',nan

# And i also dont want /5 i just need float values for better understanding 
```

```python
def handlerate(value):
    if(value== 'NEW'or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

    
df['rate']= df['rate'].apply(handlerate)
df['rate'].head(10)
```

```python
df['rate'].unique()

# As you can see now all the string values are removed
```

```python
# Lets fill the null values now

df['rate'].fillna(df['rate'].mean(), inplace= True) # inplace means i am making changes in my orignal Dataframe

df['rate'].isnull().sum()

# In the above syntax i have filled all the null values in column rate with mean

# lets check the output - we have 0 null values in column rate now
```

```python
#lets drop the null values
df.dropna(inplace=True)
df.head()
```

```python
# Lets check the unique values in cost2plates

df['Cost2Plates'].unique()

# As you see in my data that there some values having comma in between 1,800 that is stopping my value to convert into integer
```

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['Cost2Plates'] = df['Cost2Plates'].apply(handlecomma)
df['Cost2Plates'].unique()

# As you can see in the output that in the values now comma is been removed
```

```python
df['rest_type'].value_counts()
```

```python
# Lets Clean the rest_type of column

rest_type = df['rest_type'].value_counts(ascending= False)
rest_type
```

```python
rest_type_lessthan1000 = rest_type[rest_type<1000]
rest_type_lessthan1000
```

```python
# Now for all the small frequency Rest types i am going to make the cluster and name it as others

def handle_rest_type(value):
    if(value in rest_type_lessthan1000):
        return 'others'
    else:
        return value
    
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
df['location'].value_counts()
```

```python
# Cleaning the location column

location = df['location'].value_counts(ascending = False)

location_lessthan300 = location[location<300]

def handle_loaction(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
    
df['location']= df['location'].apply(handle_loaction)
df['location'].value_counts()
```

```python
# Lets clean the cusine column

df['cuisines'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending = False)

location_lessthan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in location_lessthan100):
        return 'others'
    else:
        return value
    
df['cuisines']= df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts() 

# Here also i have wrote the same function as above for loaction and have taken the value as less than100
```
